### lists
A python list is like a cell array in matlab. 

````python
# lists
list1 = list()  # creates an empty list
list1 = []  # does the same thing
list2 = [1, 2, 5]  # creates a populated list 
````

We can append entries to the end of a list using the append function. Here, the lists are object instances of a class (if you've programmed in an object-oriented language or used the classdef in matlab), and append is one of its functions.
In the case of matlab, this method of appending to the end of a vector or cell array is also possible (e.g., using vector(end+1)=2); however, while matlab discourages this behavior and encourages you to preallocate your memory, it is perfectly acceptable to use the append function in python.

````python
list1.append(12)
# the elements don't have to be of the same type:
list1.append('this is a string')
list1.append(['this', 'is', 'another', 'list'])  # lists can even contain other lists
list1.append(1234)

print(list1)  # inspect the list we created
print(len(list1))  # the len() function is analogous to the length() function in matlab

print(list2)
print(len(list2))
````

Slicing a list differs slightly from matlab because matlab indexes from 1 while python indexes from 0. Also, when slicing a range of values, in python the value corresponding to the end index is NOT included.

````python
print(list1[0])  # returns the first entry of the list
print(list1[-1])  # returns the last entry of the list

print(list1[1:2])  # this slice of list1 is a new list containing ONLY the 2nd entry of list1
print(list1[1])  # this slice returns the 2nd entry, but not as a list

print(list1[1:])  # this slices from the 2nd entry to the end
print(list1[1:-1])  # however, this does not include the last entry
````

````python
# lists can also be modified by indexed assignment:
list1[0] = 'new entry'
print(list1)
````

### dictionaries
Python dictionaries store (key, value) pairs such that specifying a key returns a value. The keys and values can take many different types, such as numbers and strings. However, a list cannot be a key (however, later we introduce tuples, which are similar to lists and can be keys).

````python
dict1 = {'spider': 8, 'dog': 4, 'human': 2, 'cat': 4}  # this dictionary specifies for each animal key the number of legs
print(dict1)
print(dict1['spider'])  # according to dict1, spiders have 8 legs

dict1['bird'] = 2  # we can add a new entry to the dictionary
print(dict1)

dict1[2] = 'this entry makes no sense'  # the keys don't have to be of the same type, but it often makes sense for them to be
print(dict1)

print(len(dict1))  # dictionaries have lengths too -- the number of (key, value) pairs

print(dict1.keys())  # to get a list of the keys, use the instance function .keys()
````

### tuples
Tuples are a lot like lists, except they are immutable -- that is, once you create a tuple, you can no longer modify it. Tuples are denoted by round brackets (while lists are denoted by square brackets).

````python
tuple1 = ('this', 'is', 'a', 'tuple')
print(tuple1)
print(len(tuple1))  # tuples have lengths

print(tuple1[0])  # you can access an element of a tuple using indexing
tuple1[0] = 'p'  # however, you cannot modify it (this will produce an error)
````

````python
dict1[tuple1] = 'm'  # tuples can act as dictionary keys
print(dict1)
````

Data structures that have multiple elements are iterable; that is, they can be traversed using a for loop.

#### Variable referencing
A crucial difference between matlab and python is the behavior of assigning a defined object to another variable. In matlab, that object is automatically copied; however, in python the new variable is merely a reference pointing to the same object. This means that if you modify the list, all the variables will update.

````python
list2 = list1  # 'list2' and 'list1' refer to the same object

# thus, if I append a new entry to list2, list1 is also updated:
list2.append('new entry')
print(list1)
````

#### list comprehension
This is a convenience functionality that doesn't really have a parallel in matlab. If your purpose is to make a list or vector or something using a for-loop, list comprehension allows you to do so in one line.

````python
[i for i in range(10) if i**2>4]  # creates a list from the iterable range(10), but only if the value squared is greater than 2
````

````python
[i for i in 'hello']  # iterates over a string to produce a list of its symbols
````

````python
[i+i for i in list1]  # adds each entry of list1 to itself
````

## numpy
Numpy is probably the most useful python library you will use -- it includes data structures to represent multidimensional arrays, such as vectors, matrices, and tensors as well as functions that operate on them. You will find pretty much all the matrix-based functions from matlab in the numpy module.

In order to use it, you will have to import it:

````python
import numpy
````

In order to use the functions provided by numpy, you would type `numpy` followed by a dot (.) followed by the function name. For example, to create a 2x2 array of ones, you would call

````python
numpy.ones((10, 10))
````

This function is directly analogous to matlab's `ones` function. In many cases, the function name is the same in both matlab and numpy. However, note that numpy's function requires a tuple or a list of dimensions, unlike matlab's function, which takes the dimensions as individual arguments. This is a common theme in numpy -- often dimensions will be passed as tuples or lists. Here is one exception: generating random numbers

````python
print(numpy.random.rand(3, 3))  # a 3x3 array of uniform random numbers between 0 and 1 
print(numpy.random.randn(3, 3))  # same thing, except sampled from a standard normal distributino
````

Here, random is a submodule of numpy (I'm not actually sure if it's called a submodule) -- it groups together a bunch of functions related to random number generation. Another example is fft -- all the fft-related functions are grouped under this submodule, such as `numpy.fft.fft`, `numpy.fft.fftshift`, and `numpy.fft.fft2`. These are directly analogous to those in matlab.

By now, you might be feeling that it's somewhat cumbersome to have to type out `numpy` every time you want to use one of its functions, whereas in matlab you can just direction use each function. There are (at least) two ways to alleviate this minor inconvenience:
```python
# option 1: rename the module

np.ones((10, 10))

# option 2: import everything

```

Option 1 is usually prefered, and the abbreviation `np` is pretty much standard. However, you can make it any name you want. Option 2 allows you to use the function without any prefixes. However, this approach is generally not used (I never use it) because you could potentially have multiple packages with the function names. In particular, tensorflow and numpy share many function names, but they are slightly different (more on this in a few weeks).

````python
import tensorflow as tf
import numpy as np
print(np.ones)
print(tf.ones)
````

One final note is that it can still be cumbersome to type out `np.fft.fft` so what I do is specifically import those functions.

````python
from numpy.fft import fft, ifft, fftshift, ifftshift, fft2, ifft2
fft
````

### some array operations
In matlab, using `*` with matrices and/or vectors interprets that as matrix multiplication and if you want elementwise products, you would need to use `.*`. In python, the latter operation is not used, and former is used to do elementwise products.

````python
A = np.ones((5, 5))
B = 2*np.ones((5, 5))
A*B
````

If you want to do matrix multiplication, you can use `np.matmul`

````python
np.matmul(A, B)
````

In general, array operations such as division and exponentiation are done elementwise, and specific np function calls do matrix operations. Note that exponentiation in python is `**`, not `^` as in matlab.

In matlab, to transpose an array, you would use `'`; in numpy you would either use `np.transpose` or follow the array with `.T`

````python
A = np.round(np.random.rand(4, 4)*10)
print(A)
print(A.T)
print(np.transpose(A))
````

### broadcasting

Broadcasting is where you can do operations between matrices whose dimensions don't match, but have the same number of dimensions (except maybe a singleton dimension). In such cases, certain dimensions of the array are tiled behind the scenes (i.e., in C) so that the dimensions match. Broadcasting exists in matlab but up until recently was cumbersome to use. You may have used the matlab function `bsxfun`. Perhaps broadcasting is better explained through examples, if it is not yet clear:

````python
# simplest case of broadcasting:
np.ones((10, 10)) + 2  # technically the dimensions don't match, but the second "array" is broadcasted to the correct shape
````

````python
# maybe you want to add a row vector to each row of a matrix:
A = np.ones((10, 10))
v = np.arange(10)  # 0, 1, 2, ... 8, 9
A+v
````

````python
# maybe you want to compute an outer product of 2 vectors
v1 = np.arange(10)
v2 = np.ones(10)

# outer product:
print(v1[None, :]*v2[:, None])  # None adds a new singleton axis

# inner product:
print(np.sum(v1*v2))
````

## function definitions and lambda functions
In matlab, you had to create a separate file for each function you defined (unless it was an anonymous function), which I found to be annoying. In python, you have function defs and lambda functions, both of which can be defined in the same file as the script or jupyter notebook.

````python
def useless_function1(x):
    return 0

def useless_function2(x):
    if x=='cat':
        return 'meow'
    elif x>3:
        return x**2
    else:
        return np.cos(x)
    
def function_that_I_havent_yet_defined():
    pass  # this is a placeholder so that the code can run
    # you can also use "pass" after any code structure that ends with a colon, after which you need to use a tab
````

````python
print(useless_function1(3.1))
print(useless_function2('cat'))
print(useless_function2(3))
````

Function definitions can be nested:

````python
def useless_function3(x):
    # these nested functions are only available to useless_function3
    def useless_helper_function1(x):
        return 2*x
    def useless_helper_function2():
        return 2*x
    # note that although x is the argument of both the function and nested functions,
    # the function will check first if x is defined locally within its scope before it
    # looks globally; examine the following behavior:
    print(2*x)  # this one uses the original input to useless_function3
    print(useless_helper_function1(3))  # this one uses x=3
    print(useless_helper_function2())  # this one uses the input to useless_function3, since a local version was not defined

useless_function3(4)
````

Lambda functions are completely analogous to matlab's anonymous function:

````python
f = lambda x: x**2
print(f(2))
print(f(3))
print(f(5.4))
````

Functions defined using lambda and functions defined using def are pretty much on equal footing. Oftentimes, another function will require a function handle (matlab's terminology), in which case you can pass either type of function.
